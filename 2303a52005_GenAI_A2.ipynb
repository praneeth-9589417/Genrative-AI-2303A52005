```python
from scipy.optimize import linprog
import numpy as np
obj=[-6,-5]
A=[[1,1],[3,2]]
b=[5,12]
bnd=[(0,float("inf")),(0,float("inf"))]
opt=linprog(c=obj,A_ub=A,b_ub=b,bounds=bnd)
chac=np.round(opt.x)
print("No.of A & No of B :", chac[0],chac[1])
```

```python
from scipy.optimize import linprog
import numpy as np
obj=[-50,-120]
A_ieq=[[7000,2000],[10,30]]
b_ieq=[7000000,1200]
A_eq = [[1,1]]
b_eq = [100]
bnd=[(0,float("inf")),(0,float("inf"))]
opt=linprog(c=obj,A_ub=A_ieq,b_ub=b_ieq,bounds=bnd,A_eq=A_eq,b_eq=b_eq,method="simplex")

Hex = np.round(opt.x)
print(f"No. of heatares for Wheat :{int(Hex[0])}")
print(f"No. of heactares for Barley :{int(Hex[1])}")
```
